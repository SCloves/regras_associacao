```python
import pandas as pd
from apyori import apriori
```

```python
dados = pd.read_csv('dados/base_pizzaria.csv', header=None)
```

```python
dados.head()
```

```python
transacoes = []
for lista in map(list, dados.values):
    transacoes.append([str(item) for item in lista])
```

```python
transacoes
```

```python
len(transacoes)
```

```python
regras = apriori(transacoes, min_suport = 0.062,
                min_confidence = 0.1,
                min_lift = 1,
                min_lenght = 2)
```

```python
resultados = list(regras)
```

```python
len(resultados)
```

```python
resultados
```

```python
resultados2 = [list(x) for x in resultados]
resultadoFormatado = []
for j in range(36):
    resultadoFormatado.append([list(x) for x in resultados2[j][2]])
```
